What was the best month in sales ...and what was the amount of money made

(i) Change the datatype  price each and quantity ordered to int

```
new_set['Quantity Ordered']=pd.to_numeric(new_set['Quantity Ordered'])
new_set['Price Each']=pd.to_numeric(new_set['Price Each'])
new_set['Order Date']=pd.to_datetime(new_set['Order Date'])
```

```
new_set.dtypes
```

```
new_set['Sales']=new_set['Quantity Ordered']*new_set['Price Each']
new_set
```

```
new_set.groupby('Month').sum()['Sales']
```

Ploting the results in a bar graph

```
import matplotlib.pyplot as plt
```

```
result = new_set.groupby('Month').sum()['Sales']
```

```
months = range (1,13)
```

```
plt.bar (months,result)
plt.xticks(months)
plt.ylabel('Sales in million USD ($)')
plt.xlabel('Month Number')

plt.show()
```

What city had the highest number of sales 

Add a city column

```
def city_finder(Address):
  return Address.split(',')[1]
def city_finder2(Address):
  return Address.split(',')[2].split(' ')[1]


new_set['City']=new_set['Purchase Address'].apply(lambda x:city_finder(x)+' '+city_finder2(x))
new_set
```

grouping now by cities 

```
sales_by_city = new_set.groupby('City').sum()
sales_by_city
```

```
city = [city for city , df in new_set.groupby(['City'])]
```

```
city
```

```
plt.bar (city,sales_by_city['Sales'])
plt.xticks(city,rotation='vertical',size=8)
plt.ylabel('Sales in million USD ($)')
plt.xlabel('Cities')

plt.show()
```

What time should we display our ads to increase the likelyhood of a customer purchasing 

```
new_set.head()
```

```
new_set['Hour']=new_set['Order Date'].dt.hour
new_set.head()
```

```
hour = [hour for hour , df in new_set.groupby(['Hour'])]
```

```
hour
```

```
sales_per_hour = new_set.groupby('Hour').sum()
sales_per_hour
```

```
plt.bar (hour,sales_per_hour['Sales'])
plt.xticks(hour,size=8)
plt.ylabel('Sales in million USD ($)')
plt.xlabel('Sales Per Hour of Day ')

plt.show()
```

```
plt.bar (hour,sales_per_hour['Quantity Ordered'])
plt.xticks(hour,size=8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Number of orders Per Hour of Day ')

plt.show()
```

line graph

```
plt.plot (hour,sales_per_hour['Quantity Ordered'])
plt.xticks(hour,size=8)
plt.ylabel('Quantity Ordered')
plt.grid()
plt.xlabel('Number of orders Per Hour of Day ')

plt.show()
```
